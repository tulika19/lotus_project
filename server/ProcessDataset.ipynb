```python
#import the pandas library

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import csv

# read in the csv file as a pandas dataframe
print('WORKBANK DATASET LOAD AND DROP NA')
df_worldBank = pd.read_csv('world_bank_dataset_1.csv')

# explore the data
df_worldBank.info()

df_worldBank.shape

# checking our dataframe to see if the mappings have worked
df_worldBank.head()

df_worldBank_new=df_worldBank.dropna()
#print(df_worldBank_new)
# read in the csv file as a pandas dataframe
#print('LOTUS DATASET LOAD AND DROP NA')

df_lotus_dataset = pd.read_csv('Data_lotus_dataset.csv')
df_new_lotus= df_lotus_dataset.dropna()
#print(df_new_lotus)

# explore the data
df_new_lotus.info()

df_new_lotus.shape

# checking our dataframe to see if the mappings have worked
df_new_lotus.head()

merged = pd.merge(df_worldBank_new,df_new_lotus, left_on = 'Country Name', right_on = 'Country', how = 'inner')

print(merged.head())
#print(type(merged))

# count the unique values in the Country column
#print(merged.Country.value_counts())

#merged.to_csv(path_or_buf="D:\lotus_project\lotus_project\server\file.csv")

print('AFTER MERGED CHECK POVERTY PER COUNTRY')

def calculate_poverty_rate(merged):
    poverty_rate = merged.groupby(['Country'])['Poverty'].mean().reset_index()
    print(poverty_rate)
    poverty_rate.plot()
    plt.show()
def calculate_Education_rate(merged) : 
    education_rate = merged.groupby(['Country'])['Education'].mean().reset_index()
    print(education_rate)
    education_rate.plot(kind = 'bar', x = 'Country', y = 'Education')
    plt.show()
    
    
calculate_poverty_rate(merged)
calculate_Education_rate(merged)

```
